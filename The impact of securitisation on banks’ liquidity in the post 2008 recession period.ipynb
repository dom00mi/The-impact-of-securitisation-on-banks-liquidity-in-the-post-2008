# The impact of securitisation on banks’ liquidity in the field of global securitisation in the post 2008 recession period

### Introduction:

When thinking of securitization, the common strand in anyone's memory is the 2007 subprime mortgage crisis. The bad reputation of securitization comes from its ruinous effect on the 2008 financial crisis. Briefly, securitization means the process in which various kinds of contractual debts such as residential mortgages or auto loans are pooled so that they can be sliced up and repackaged into new securities. This report aims to truly understand the key role and the utmost implications of securitization in the banking world, its effect on a bank's liquidity and global economy.


### Securitization: a cornerstone in modern banking

Securitization finds its roots in the late 17th Century when the first mercantilist corporations acted as vehicles of sovereign debt for the British Empire1. In truth, securitization has started playing a pivotal role in the financial landscape only after 1970s and it was boosted by the high demand for the flourishing housing market. The growth of this last was further promoted by political decisions such as the Bush politics of homeownership. The US economy has seen an enormous expansion of the securitization market, especially securitized home mortgages growing enormously from 1976 to the end of 2007 as shown by Federal Reserve “Flow of Funds Accounts of the United States.




### Implications of securitization

First, it represents a new source of financing banks' investments, in that by securitizing outstanding loans, banks can have other funding for providing new loans. Securitization allows banks to transform very illiquid and hard-to-sell assets such as mortgages into marketable and tradable securities. With their appealing nature, these loans can be sold all around the World and money can flow from the regions with excess deposits to the regions with the unsatisfied loan demand. From this, it originates the global nature of securitization.

Second, this process provides an opportunity for banks to hold more diversified loan portfolios and therefore protecting them against local economic shocks while at the same time making them more prone to the global economic declines.

However, a downside of securitization imperils fundamental screening and monitoring roles of banks, by allowing them to easily move assets off their balance sheets.

### Securitization and on-balance sheet liquidity

Within the securitization, thanks to creating liquid financial instruments out of assets that could be too cumbersome or expensive to sell individually, if a bank can securitize the existing loans as easily as it can turn liquid assets into cash, such a bank is unlikely to hold liquid assets.

### The utmost meaning of liquidity

As a measure of liquidity, we will use the "Liquidity Mismatch Index" which gauges the mismatch between the market liquidity of assets and the funding liquidity of liabilities.

Generally, for assets $(a^i_{t,k})$, liabilities $(l^i_{t,k'})$ with respect to an entity $i$ at a given time $t$, the $LMI^i_t$
is given by
\begin{equation}
LMI^i_t=\sum_{k} \lambda_{t,a,k} a^i_{t,k} + \sum_{k'} \lambda l^i_{t,k'}	
\end{equation}

The LMI is measured at the bank level, but it will also be appealing to aggregate the LMI across the banking sector.


According to , let us define $[LMI]^−_t$ which measures the aggregate liquidity vulnerability of the banking system as,
\begin{equation}
[LMI]^− =\sum_{i}min[LMI_i ,0] 
\end{equation}

The LMI-minus is the sum across only those banks with a negative LMI, and thus measures the liquidity shortfall in the systemic event that every bank that is susceptible to a run, suffers that run.

Their low liquidity affects their financial health. As a result, when borrowers' defaults and consequent huge losses for investors were occurring, banking stability was threatened. To prevent a lack of liquidity, the Federal Reserve provided over $1.5 trillion of liquidity support across its many liquidity facilities.

### Conclusion:

The financial crisis was a cornerstone in allowing us to truly understand the utmost meaning of securitisation. The prompt actions played by regulators and the Federal Reserve were of great importance to both re-establish liquidity conditions and change the game's rule. Despite its disastrous effects on the economy, the convergence between securitisation and financial crisis is highly linked to factors such as creditworthiness, products' characteristics and regulation. While massively misused during the 2000s, a well regulated and efficiently used securitization is the pillar of a new global financial landscape. The toxicity of those assets is to be re-examined in the context of a very clear, simple and transparent securitisations which should have a place in modern banking. Its purpose is to free up funding for banks, allowing them to sell packages of mortgage loans and lend out the proceeds to more customers. Those loans could then be sold on, and the cycle could start again5. Ultimately, securitization could make the financial system less risky by giving banks more sources of funding as told by Mr Rule, the Bank of England’s executive director of prudential policy. This reflection on securitization makes us rethink about this term in the view of a new financial landscape even if among its benefits, there are still dangers to be aware of.
